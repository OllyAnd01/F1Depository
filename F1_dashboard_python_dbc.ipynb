```python
#!pip install dash
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objects as go
import numpy as np
import pandas as pd
from plotly.subplots import make_subplots
```

```python
#df = pd.read_csv(r"C:\Users\ollya\OneDrive\Dashboard Project\archive\races.csv")
df = pd.read_csv(r"C:\Users\PC\OneDrive\Dashboard Project\archive\races.csv")


df = df[df['year'] == 2024]
df = df[['raceId', 'year', 'name', 'date', 'circuitId']]
df['name'] = df['name'].str.replace('Grand Prix', '').str.strip()
df
```

```python
#df2 = pd.read_csv(r"C:\Users\ollya\OneDrive\Dashboard Project\archive\results.csv")
df2 = pd.read_csv(r"C:\Users\PC\OneDrive\Dashboard Project\archive\results.csv")

#df2 = df2[df2['raceId'] == 1132]
x = df2['driverId']
x
```

```python
#df3 = pd.read_csv(r"C:\Users\ollya\OneDrive\Dashboard Project\archive\drivers.csv")
df3 = pd.read_csv(r"C:\Users\PC\OneDrive\Dashboard Project\archive\drivers.csv")

df3 = df3.drop(columns=['number'])
df3 = df3[(df3.driverId.isin(x))]
df3
```

```python
df4 = df2.merge(df3, on='driverId', how='left')
df4['fullname'] = df4['forename'] + ' ' + df4['surname']
df4 
```

```python
df5 = df4.merge(df, on='raceId', how='left')
df5 = df5[df5['year'] == 2024]
df5.columns
```

```python
df6 = pd.read_csv(r"C:\Users\PC\OneDrive\Dashboard Project\archive\constructors.csv")
df6 = df6[df6['constructorId'].isin(df5['constructorId'])]
df6 =df6.rename(columns={'name' : 'constructor_name'})
df6
```

```python
df7 = df5.merge(df6, on='constructorId', how='left')
df7.columns
```

```python
df_longlat = pd.read_csv(r"C:\Users\PC\OneDrive\Dashboard Project\archive\circuits.csv")
df_longlat = df_longlat.rename(columns={'name' : 'circuit_name'})

df7 = df7.merge(df_longlat, on='circuitId', how = 'left' )
df7
```

```python
print(df7.position.unique())
df7.position = df7['position'].replace(['\\N', ''])
df7.position = df7['position'].astype(int)
```

```python
df7['constructorRef'] = df7['constructorRef'].str.replace("_", " ")
df7.constructorRef = df7.constructorRef.str.title()
```

```python

df7['milliseconds'] = df7['milliseconds'].replace(r'\N', '')
print(df7['milliseconds'])
df7['laptime'] = pd.to_datetime(df7['milliseconds'], unit='ms').dt.time
df7['laptime']
```
