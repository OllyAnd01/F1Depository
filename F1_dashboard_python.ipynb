```python
#!pip install dash
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objects as go
import numpy as np
import pandas as pd
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv(r"C:\Users\PC\Downloads\archive\races.csv")

df = df[df['year'] == 2024]
df = df[['raceId', 'year', 'name', 'date']]
df
```

```python
df2 = pd.read_csv(r"C:\Users\PC\Downloads\archive\results.csv")
#df2 = df2[df2['raceId'] == 1132]
x = df2['driverId']
x
```

```python
df3 = pd.read_csv(r"C:\Users\PC\Downloads\archive\drivers.csv")
df3 = df3.drop(columns=['number'])
df3 = df3[(df3.driverId.isin(x))]
df3
```

```python
df4 = df2.merge(df3, on='driverId', how='left')
df4['fullname'] = df4['forename'] + ' ' + df4['surname']
df4 
```

```python
df5 = df4.merge(df, on='raceId', how='left')
df5 = df5[df5['year'] == 2024]
df5.columns
```

```python
df5.dtypes
```

```python
print(df5.position.unique())
df5.position = df5['position'].replace(['\\N', ''])
df5.position = df5['position'].astype(int)
```

```python
df5.to_excel(r"C:\Users\PC\Downloads\F1_dashboard_data.xlsx", index=False)
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

app = Dash(__name__)

app.layout = html.Div([
    html.H4('F1 Dashboard'),

    dcc.Dropdown(
        id = "dropdown",
        options = [{'label': i, 'value': i} for i in df5['fullname'].unique()],
        value = df5['fullname'].unique()[0],
        clearable = False,
    ),
    dcc.Graph(id='graph'),

    ##dcc.Graph(id='graph2')
])

    

@app.callback(
    Output('graph', 'figure'),
    Input('dropdown', 'value'))
def update_bar_chart(driver):

    df = df5[df5['fullname'] == driver]
    df['moving_avg'] = df['position'].rolling(window=3, min_periods=1).mean()
    df['cumulative_points'] = df['points'].cumsum()
    df['date'] = pd.to_datetime(df['date'])
    df = df.sort_values(by='date')


    fig = make_subplots(rows =1, cols=2)
    #go.Figure()
    
    fig.add_trace(go.Bar(
        x = df['name'],
        y = df['position'],
        name = 'Race Position',
        marker_color = 'blue',
    ),row=1, col=1)

    fig.add_trace(go.Scatter(
        x = df['name'],
        y = df['moving_avg'],
        mode = 'lines+markers',
        name = 'Moving Average',
        line = dict(color='red', width=2),
    ),row=1, col=1)

    fig.update_layout(
        title=f'Position of {driver} in 2024 Season',
        xaxis_title = 'Grand Prix',
        barmode = 'overlay'
    )

    fig.add_trace(go.Bar(
        x = df['name'],
        y = df['points'],
        name = 'Points',
        marker_color = 'green',
    ), row=1, col=2)

    fig.add_trace(go.Scatter(
        x = df['name'],
        y = df['cumulative_points'],
        mode = 'lines+markers',
        name = 'Cumulative Points',
        line = dict(color='orange', width=2),
    ), row =1, col =2)
    

    return fig


if __name__ == '__main__':
    app.run(debug=True)
```
